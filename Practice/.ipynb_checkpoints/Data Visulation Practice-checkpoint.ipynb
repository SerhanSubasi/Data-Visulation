```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  #mpl şeklinde de karşımıza çıkar.
```

##  Plot
### Çizgisel grafiktir.

```python
x = np.linspace(0,10,20)
y = 2 * x
```

```python
plt.figure(figsize = (4,4))   #figsize yazmasak da olur, yazarsak boyutu manuel biz ayarlarız.
plt.plot(x,y)  #plot ile çizgisel bir grafik oluştururuz.

plt.show()  #plt.show ile aşağıda bilgi yazılarının çıkmamasını sağlarız.
```

## Scatter

### Dağılım grafiğidir.

```python
x = np.array([4, 6, 8, 7, 7, 15, 5, 9, 4,
              11, 13, 8, 7, 8 ,11, 9, 12,
              15, 17, 17, 19, 18 ])
y = np.array([89, 76, 77, 78, 101, 76, 93,
             77, 84, 68, 67, 75, 76, 88,
             84, 93, 97, 101, 102, 113, 114, 110])
```

```python
plt.figure(figsize=(4,4))
plt.scatter(x,y)
plt.show()
```

## Bar Plot
### Barların olduğu, kategoriksel verilerde kullanılan grafiktir.

```python
x = np.array(["A","B","C","D","E"])
y= np.array([5,11,3,14,8])
```

```python
plt.figure(figsize=(4,4))
plt.bar(x,y)
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.bar(x,y, color="r")
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.barh(x,y) #h ekleyerek horizonal yani yatay da yapabiliyoruz.
plt.show()
```

## Histogram
### Bitişik grafiktir

```python
x = np.random.normal(170,10,250)  #ortalama, std, adet
```

```python
plt.figure(figsize=(4,4))
plt.hist(x)
plt.show()
```

## Box Plot
### Outlier bulabildiğimiz grafik türüdür.

```python
plt.figure(figsize=(4,4))
plt.boxplot(x)
plt.show()
```

## Pie Chart
### Pasta dilimleri grafiğidir, labels ile çalışır.

```python
x = ["Kategori 1", "Kategori 2", "Kategori 3", "Kategori 4"]
sizes = [15,30,45,10]
```

```python
plt.figure(figsize=(5,5))
plt.pie(sizes, labels=x)
plt.show()
```

# Multi Data ve Legend 

```python
age = np.arange(24,44,2.5)
salary = np.arange(40000,76000,5000)
salary_2 = [45372, 46852, 69401, 56790, 73000, 75000, 80000, 83454]
```

```python
plt.figure(figsize= (6,4))
plt.plot(age,salary, "b:", label="Salary 1")    # label ile legend'in etiketlemesini yaptık.
plt.plot(age,salary_2, "r", label="Salary 2")   # label ile Legend'in etiketlemesini yaptık.

plt.xlabel("Yas")
plt.ylabel("Maas")
plt.title(label = "Yasa göre maas grafiği")

plt.legend()  #legend, etiketleri grafiğe ekler.
plt.show()
```

```python
plt.figure(figsize= (6,4))
plt.plot(age,salary, "b:", label="Salary 1")    
plt.plot(age,salary_2, "r", label="Salary 2")   

plt.xlabel("Yas")
plt.ylabel("Maas")
plt.title(label = "Yasa göre maas grafiği")

plt.legend(loc="lower right")   # loc = "lower right" ile legend'in lokasyonunu belirttik.
plt.show()
```

```python
plt.figure(figsize= (6,4))
plt.plot(age,salary, "b:", label="Salary 1")    
plt.plot(age,salary_2, "r", label="Salary 2")   

plt.xlabel("Yas")
plt.ylabel("Maas")
plt.title(label = "Yasa göre maas grafiği")

plt.legend(loc=5)   # loc = 5 ile legend'in lokasyonunu sayı girerek de belirtebiliriz.
plt.show()
```

```python
plt.figure(figsize= (6,4))
plt.plot(age,salary, "b:", label="Salary 1")    
plt.plot(age,salary_2, "r", label="Salary 2")   

plt.xlabel("Yas")
plt.ylabel("Maas")
plt.title(label = "Yasa göre maas grafiği")

plt.legend(bbox_to_anchor=(0,0,0,0))   # Legend'i grafik dışında görmek istiyorsak bu kodu kullanabiliriz.
plt.show()
```

# Figure ve Axes Renklendirmesi

```python
plt.figure(figsize= (6,4), facecolor="Yellow")
ax = plt.axes()
ax.set_facecolor("xkcd:mint green")

plt.plot(age,salary, "b:", label= "Salary 1", lw= 4)
plt.plot(age,salary_2, "r", label= "Salary 2", ls= "-")

plt.xlabel("Yas")
plt.ylabel("Maas")
plt.title(label="Yasa göre maas grafiği")

plt.legend(loc="lower right")
plt.show()
```

# Functional Method

```python
plt.subplot(2,1,1)     # 2 satır 1 sütundan oluşan grafiğin 1.si
plt.plot(age, salary, "b")

plt.subplot(2,1,2)     # 2 satır 1 sütundan oluşan grafiğin 2.si
plt.plot(age,salary_2, "r")

plt.tight_layout()   #Bu kod grafiklerin birbirine girmemesi için yazılır
plt.show()
```

# Object Oriented Method (OOM)

```python
fig , ax = plt.subplots(2,3)

ax[0,0].plot(age,salary, "b")
ax[0,1].plot(age,salary_2, "r")
ax[0,2].plot(age,salary, "g")

ax[1,0].plot(age,salary_2, "yellow")
ax[1,1].plot(age,salary, "brown")
ax[1,2].plot(age,salary_2,"purple")

plt.tight_layout()   #Bu kod grafiklerin birbirine girmemesi için yazılır
plt.show()
```
