## Count Plot

```python
sns.countplot(data = tips, x = "day");
```

```python
fig, ax = plt.subplots()

ax = sns.countplot(data = tips, x = "day")

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+0.2))
```

p.get_height() = y dzlemindeki denk gelen ykseltideki rakam en ste yaz

p.get_x()+0.3 = yatay dzlemde yerini belirtiyor (y ekseni)

p.get_height()+0.5 = dikey dzlemde yerini belirtiyor (y ekseni)

```python
fig, ax = plt.subplots(figsize = (10, 5))
ax_tip = sns.countplot(data = tips, x = "day", hue = "sex")
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.15, p.get_height()+1))
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
ax_tip = sns.countplot(data = tips, x = "day", hue = "sex")
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.15, p.get_height()+1))
    ax.text(p.get_x()+0.07, p.get_height()-4, "%"+str(round(p.get_height()/tips.day.count()*100, 2)));
```

## Bar Plot

izgileri gven araldr, confidence interval yani.

```python
sns.barplot(data = tips, x = "sex", y = "tip"); #meandefault= np.mean
```

```python
sns.barplot(x = "sex", y = "tip", data = tips, estimator = np.sum); #sum
```

```python
sns.barplot(x = "sex", y = "tip", data = tips, estimator = np.mean, ci = "sd");
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize= (10, 5))
ax[0] = sns.barplot(x = "day", y = "total_bill", data = tips, ax = ax[0])
ax[1] = sns.countplot(x = "day", data = tips, ax = ax[1])
# barplotta default estimator deeri np.mean olarak alnr
```

bu iki plot tr farkl bilgiler sunar:

Barplot

sns.barplot kullanlarak izilen grafie gre, x eksenindeki her bir day (gn) iin total_bill (toplam hesap) stununun ortalama deeri gsterilir. Yani bu plot, belirli gnlerdeki ortalama toplam hesap miktarn gsterir. barplot, y eksenindeki deerlerin (bu rnekte total_bill) istatistiksel bir lmn (rnein ortalama, medyan vb.) grselletirir. Bu durumda, her bir gn iin toplam hesabn ortalama deeri hesaplanr ve barlarn ykseklii bu ortalamay temsil eder. Hatalarn (gven aralklar) gsterilip gsterilmeyecei ci parametresi ile ayarlanabilir, bu sayede veri setindeki varyans hakknda da bilgi verilmi olur

Countplot

sns.countplot kullanlarak izilen grafie gre, x ekseninde yer alan her bir day iin, o gn veri erevesinde ka tane kayt olduu gsterilir. Yani bu plot, veri erevesindeki belirli gnlerdeki kayt (satr) saysn sayar ve grselletirir. countplot, herhangi bir saysal deer hesaplamadan, kategorik verinin frekansn (saysn) gsterir. Her bir bar, o kategorinin veri erevesinde ka kez getiini gsterir. Bu plot, y ekseninde zel bir hesaplama yapmaz; sadece belirli bir kategorideki gzlem saysn gsterir.

zet Barplot, kategorik veriye karlk gelen saysal bir deerin (rnein, ortalama) grselletirilmesi iin kullanlrken, Countplot, kategorik verinin kendi iindeki frekansnn (saysnn) grselletirilmesi iin kullanlr.
