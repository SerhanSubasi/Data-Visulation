
## Data

Genellikle "miles per gallon" yani galon başına mil olarak ifade edilen yakıt tüketim verimliliği üzerine çeşitli analizler yapmak için kullanılır. Veri seti, 1980'lerin sonlarına kadar Amerika Birleşik Devletleri'nde üretilen çeşitli otomobillerin özelliklerini içerir.

mpg veri setinde tipik olarak şu bilgiler bulunur:

**mpg**: Araçların galon başına kat edebildiği mil (yakıt verimliliği) olarak ölçülür.

**cylinders**: Motorun silindir sayısı.

**displacement**: Motor hacmi, genellikle inç küp (ci) veya santimetre küp (cc) olarak ifade edilir.

**horsepower**: Motor gücü, beygir gücü (hp) cinsinden.

**weight**: Araç ağırlığı, pound veya kilogram cinsinden.

**acceleration**: 0'dan 60 mil/saat hıza çıkış süresi, saniye cinsinden.

**model_year**: Araç model yılı.

**origin**: Araç üretim ülkesi.

**name**: Araç model adı.

Bu veri seti, otomobillerin yakıt verimliliği ile bu özellikler arasındaki ilişkileri analiz etmek için kullanılır. Örneğin, araştırmacılar motor hacmi, ağırlık ve silindir sayısı gibi faktörlerin mpg üzerindeki etkisini inceleyebilirler.

p.get_height() = y düzlemindeki denk gelen yükseltideki rakamı en üste yaz

## Bar and Count Plots

bu iki plot türü farklı bilgiler sunar:

Barplot

sns.barplot kullanılarak çizilen grafiğe göre, x eksenindeki her bir day (gün) için total_bill (toplam hesap) sütununun ortalama değeri gösterilir. Yani bu plot, belirli günlerdeki ortalama toplam hesap miktarını gösterir. barplot, y eksenindeki değerlerin (bu örnekte total_bill) istatistiksel bir ölçümünü (örneğin ortalama, medyan vb.) görselleştirir. Bu durumda, her bir gün için toplam hesabın ortalama değeri hesaplanır ve barların yüksekliği bu ortalamayı temsil eder. Hataların (güven aralıkları) gösterilip gösterilmeyeceği ci parametresi ile ayarlanabilir, bu sayede veri setindeki varyans hakkında da bilgi verilmiş olur

Countplot

sns.countplot kullanılarak çizilen grafiğe göre, x ekseninde yer alan her bir day için, o gün veri çerçevesinde kaç tane kayıt olduğu gösterilir. Yani bu plot, veri çerçevesindeki belirli günlerdeki kayıt (satır) sayısını sayar ve görselleştirir. countplot, herhangi bir sayısal değer hesaplamadan, kategorik verinin frekansını (sayısını) gösterir. Her bir bar, o kategorinin veri çerçevesinde kaç kez geçtiğini gösterir. Bu plot, y ekseninde özel bir hesaplama yapmaz; sadece belirli bir kategorideki gözlem sayısını gösterir.

Özet Barplot, kategorik veriye karşılık gelen sayısal bir değerin (örneğin, ortalama) görselleştirilmesi için kullanılırken, Countplot, kategorik verinin kendi içindeki frekansının (sayısının) görselleştirilmesi için kullanılır.
