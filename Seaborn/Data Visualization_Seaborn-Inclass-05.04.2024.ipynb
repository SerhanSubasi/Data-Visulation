```python
import pandas as pd

import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt   

import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

## Seaborn Data Sets

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

Seaborn kütüphanesinde bulunan 'tips' veri seti, restoranlarda bahşiş verme davranışı ile ilgili verileri içerir. Bu veri seti, farklı öğünlerde, farklı günlerde ve farklı cinsiyet/kişi sayıları için alınan bahşiş miktarlarını içermektedir. Aşağıda, bu veri setindeki sütunların açıklamaları bulunmaktadır:

1. **total_bill**: Toplam hesap miktarı (ABD doları cinsinden).
2. **tip**: Bahşiş miktarı (ABD doları cinsinden).
3. **sex**: Hesabı ödeyen kişinin cinsiyeti (Erkek/Kadın).
4. **smoker**: Grubun içinde sigara içen olup olmadığı (Evet/Hayır).
5. **day**: Hesabın ödendiği gün (Perşembe, Cuma, Cumartesi, Pazar).
6. **time**: Hesabın ödendiği öğün zamanı (Öğle yemeği/Akşam yemeği).
7. **size**: Grubun kişi sayısı.

Bu veri seti, bahşiş verme alışkanlıklarını çeşitli açılardan incelemek için kullanılabilir.

## Scatter Plot

```python
sns.scatterplot(data=tips, x="total_bill", y="tip");
```

## Reg Plot

```python
sns.regplot(data=tips, x="total_bill", y="tip");
```

## Figsize

```python
plt.figure(figsize=(7,4))
sns.scatterplot(x=tips["total_bill"], y=tips["tip"])

plt.show()
```

## Hue

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "smoker") # hue must be categorical
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "day") # hue must be categorical
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "sex") # hue must be categorical
plt.show()
```

## Style

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", style="sex")
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", style="sex", hue="day")
plt.show()
```

```python
plt.figure(figsize=(10, 5)) 
sns.scatterplot(data=tips, x="total_bill", y="tip", style="sex", markers={"Male": "s", "Female": "D"});
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 5))
sns.scatterplot(x = "total_bill", y = "tip", data = tips, style = "sex", ax = ax[0])
sns.scatterplot(x = "total_bill", y = "tip", data = tips, hue = "smoker", ax = ax[1]);

plt.show()
```

## Palette

https://seaborn.pydata.org/tutorial/color_palettes.html #seaborn renk sayfası

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="total_bill", data = tips, y = "tip", hue="sex", palette= "plasma")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="total_bill", data = tips, y = "tip", hue="sex", palette= "viridis")
plt.show()
```

```python
sns.set_palette("pastel")
# veya özel bir palet için
custom_palette = sns.color_palette(["#9B59B6", "#3498DB", "#95A5A6", "#E74C3C", "#34495E", "#2ECC71"])
sns.set_palette(custom_palette)
```

## Alpha - Yoğunluk

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="total_bill", data = tips, y = "tip", alpha=0.3) #alpha, noktaların şeffaflığını belirtir 0-1 arası
plt.show()
```

```python
# yukarıdakş başlıkların hepsini uygulayalım

plt.figure(figsize=(10,5))
sns.scatterplot(x="total_bill", data = tips, y = "tip", style="sex", hue="day", palette="plasma", alpha=0.5)
plt.show()
```

## S - Size

Grafikteki data noktalarının boyutunu pixel olarak ayarlamak için S parametresini kullanacağız

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="total_bill", data = tips, y = "tip", style="sex", hue="day", palette="plasma", alpha=0.7, s=300) #300 piksel
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="total_bill", data = tips, y = "tip", style="sex", hue="day", palette="cubehelix", alpha=0.7, s=150) 
plt.show()
```

## Distribution Plot

```python
sns.displot(data=tips, x="total_bill")
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", kde=True) #kernel çizgisini True yaptık
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", bins=30) #barların sayısını artırır.
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", bins=50, kde=True) #kde şimdi daha gerekli, dağılım çok şey ifade etmiyor çünkü şu an.
plt.show()
```

```python
sns.displot(data = tips, x="total_bill", bins=tips.total_bill.count(), kde=True); #data sayısı kadar çizgi 
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,6))

sns.scatterplot( x = "total_bill", y="tip", data=tips, s=150, hue="sex", ax= ax[0])
sns.histplot(x = "total_bill", data= tips, kde= True, ax = ax[1])
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", kind="hist")
sns.displot(data=tips, x="total_bill", kind="kde")
plt.show()
```

```python
sns.displot(x="total_bill", data = tips, bins = 244, kde=True, aspect = 3) #aspect, grafiğin genişliğinin yüksekliğinden
# kaç katı olması gerektiğini belirtir.
plt.show()
```

## RugPlot

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 5))
sns.rugplot(data=tips, x = "total_bill", height= 0.8, ax = ax[0])
sns.histplot(data=tips, x = "tip", kde = True, ax = ax[1]);
```
