```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt#mpl de olabilir.
```

### Matplotlib

### Line

```python
# Data oluturma

x = np.linspace(0, 10, 20)

y = 2 * x
```

```python
# Grafik oluturma
plt.figure(figsize = (6, 6))
plt.plot(x, y)

plt.show() #Grafii gster demek, aadaki yazy kaldrr.
```

```python
plt.figure(figsize = (6, 6))
plt.plot(x, y); #;  de show yerine kullanlabilir
```

### Scatter

```python
plt.figure(figsize= (8,4))
plt.scatter(x,y)

plt.show()
```

```python
x = np.array([4, 6, 8, 7, 7, 15, 5, 9, 4,
              11, 13, 8, 7, 8 ,11, 9, 12,
              15, 17, 17, 19, 18 ])
y = np.array([89, 76, 77, 78, 101, 76, 93,
             77, 84, 68, 67, 75, 76, 88,
             84, 93, 97, 101, 102, 113, 114, 110])

plt.figure(figsize=(8,4))
plt.scatter(x,y)

plt.show()
```

### Bar Plot

```python
x = np.array(["A", "B", "C", "D", "E"])
y = np.array([5, 11, 3, 14, 8])

plt.bar(x, y)
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.barh(x,y, color="red")

plt.show()
```

```python
plt.barh(x,y, color="aqua"); #sonuna h ekleyince horizonal yani yatay olur.
```

### Histogram

```python
x = np.random.normal(170,10,250) #ortalama, std, adet = 250 tane ortalamas 170 ve std'si 10 olan grafik rettik.

plt.figure(figsize=(4,3))
plt.hist(x)

plt.show()
```

### Box Plot

```python
plt.figure(figsize= (10,6))
plt.boxplot(x);
```

### Pie Chart

```python
#Labels
x = ['Kategori 1', 'Kategori 2', 'Kategori 3', 'Kategori 4']

import matplotlib.pyplot as plt

# Veri
sizes = [15, 30, 45, 10]

# Pasta grafiini olutur
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=x)

# Gster
plt.axis('equal')
plt.show()
```
